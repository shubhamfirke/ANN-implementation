```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/sagnikghoshcr7/Bank-Customer-Churn-Prediction/master/Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(columns=['RowNumber', 'CustomerId', 'Surname'],inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Geography'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
df=pd.get_dummies(df, columns=['Geography','Gender'],drop_first=True)
```

```python
df.head()
```

```python
x=df.drop(["Exited"],axis=1)
```

```python
y=df['Exited']
```

```python
import tensorflow
from tensorflow import keras
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
len(x.columns)
```

```python
model=Sequential()
model.add(Dense(11, activation="sigmoid", input_dim=11)) # input laer
model.add(Dense(11, activation="sigmoid")) # hidden layer
model.add(Dense(1, activation="sigmoid")) # output layer
```

```python
model.summary()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scale=scaler.fit_transform(x_train)
x_test_scale=scaler.fit_transform(x_test)
```

```python
 model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train_scale,y_train, epochs=100, verbose=1)
```

```python
y_pred=model.predict(x_test_scale)
```

```python
y_pred
```

```python
y_pred=y_pred.argmax(axis=-1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
